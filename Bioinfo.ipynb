# what are the information we get in a fasta file

The information that can be found in a FASTA file includes:

    Header line: This line provides descriptive information about the sequence, such as the sequence identifier, description, organism, gene name, or any other relevant details. The specific information included in the header line varies depending on the source and purpose of the sequence data.

    Sequence: The sequence itself represents the actual nucleotide or protein sequence. It consists of a series of characters that can include letters representing the nucleotides (A, T, G, C for DNA sequences) or amino acids (single-letter codes for proteins, such as A, R, N, etc.).

# To read multiple FASTA files contained within a single file using Biopython

i. you can utilize the SeqIO.parse() function from the Bio.SeqIO module. 
ii. In this example, you need to replace "path/to/your/file.fasta" with the actual path to your file containing the multiple FASTA sequences. 
iii. The SeqIO.parse() function reads the sequences from the file, and then you can iterate over them using a for loop. 
iv. Inside the loop, you can access various properties of each sequence, such as the ID, description, and sequence itself. 



```python
from Bio import SeqIO

# Define the path to your file containing multiple FASTA sequences
file_path = "/home/charith/Desktop/biopython/multiple.fasta"

# Use SeqIO.parse() to read the sequences
sequences = SeqIO.parse(file_path, "fasta")

# Iterate over the sequences and perform desired operations
for sequence in sequences:
    # Access sequence information
    sequence_id = sequence.id
    sequence_description = sequence.description
    sequence_sequence = sequence.seq
    
    # Perform scientific analysis or any other desired operations
    # For example, you can calculate the length of the sequence
    sequence_length = len(sequence)
    
    # Print sequence information
    print("ID:", sequence_id)
    print("Description:", sequence_description)
    print("Sequence:", sequence_sequence)
    print("Length:", sequence_length)
    print("-------------")


```

# To store the names, sequences, and lengths of sequences in a tabular form using pandas and save it as a CSV file,

The code will extract the names, sequences, and lengths of the sequences and store them in separate lists. Then, a DataFrame is created using the pandas library, where the lists are used as columns. Finally, the DataFrame is saved as a CSV file using the to_csv() function, specifying the desired output file path ('/home/charith/Desktop/output.csv' in this example).

```python
import pandas as pd
from Bio import SeqIO

# Define the path to your file containing multiple FASTA sequences
file_path = "/home/charith/Desktop/biopython/multiple.fasta"

# Use SeqIO.parse() to read the sequences
sequences = SeqIO.parse(file_path, "fasta")

# Create empty lists to store the data
names = []
sequences_list = []
lengths = []

# Iterate over the sequences and extract the information
for sequence in sequences:
    # Access sequence information
    sequence_id = sequence.id
    sequence_sequence = str(sequence.seq)
    
    # Calculate the length of the sequence
    sequence_length = len(sequence)
    
    # Append the data to the lists
    names.append(sequence_id)
    sequences_list.append(sequence_sequence)
    lengths.append(sequence_length)

# Create a DataFrame using the lists
data = pd.DataFrame({'Name': names, 'Sequence': sequences_list, 'Length': lengths})

# Define the output file path
output_path = '/home/charith/Desktop/biopython/output.csv'

# Save the DataFrame as a CSV file
data.to_csv(output_path, index=False)

# Print a message indicating the successful creation of the CSV file
print("CSV file saved at:", output_path)
```

# what are the information we get in a genbank file

In a GenBank file, you can find various pieces of information related to a biological sequence. Here are some common pieces of information typically included in a GenBank file:

1. Locus: This section provides general information about the sequence, including the locus name, sequence length, molecule type, and other metadata.

2. Features: This section describes the features annotated on the sequence, such as genes, coding sequences (CDS), regulatory regions, repeat regions, etc. Each feature is typically defined by its location (start and end coordinates) on the sequence and associated qualifiers.

3. Origin: This section contains the actual sequence data, typically represented as a string of nucleotides (A, T, G, C) for DNA sequences or a string of bases (A, U, G, C) for RNA sequences.

4. References: This section provides references to scientific articles or other sources that describe the sequence or related information.

5. Keywords: This section includes keywords or terms that describe the sequence and provide additional information about its characteristics or relevance.

6. Organism: This section specifies the scientific name of the organism from which the sequence was derived.

7. Version: GenBank files often include a version number or identifier for the sequence entry, indicating any revisions or updates made to the sequence.

8. Other miscellaneous information: GenBank files may contain additional fields or information specific to the sequence entry, such as authors, taxonomy, sequence history, etc.

It's important to note that the specific structure and contents of a GenBank file may vary depending on the database or source from which it was obtained. The above information represents common elements typically found in GenBank files, but additional information may be present depending on the level of annotation and metadata available for a particular sequence.

# Pairwise Sequence Alignment

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
from Bio.Seq import Seq

seq1 = Seq('ATGCGCGCGCGCGCGCGGCGCG')
seq2 = Seq('ATGCATGCATGC')

print('Length of SEQ1: ',len(seq1))
print('Length of SEQ2: ',len(seq2))
print('.................................................................................................')

ali = pairwise2.align.globalxx(seq1,seq2)
for i in ali:
    print(i)
    print('..............................................................................................')
    
    #Bio.pairwise2 module provides a formatting method, format_alignment to better visualize the result
    print(format_alignment(*i))
```

# Biopython to perform MSA, construct a phylogenetic tree using the UPGMA method, and visualize the tree. 

```python
from Bio import Phylo
from Bio.Align.Applications import MuscleCommandline
from Bio import AlignIO
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor, DistanceCalculator

file = "/home/charith/Desktop/multiple.fasta"

# Perform sequence alignment using MUSCLE
aligned_file = "/home/charith/Desktop/aligned_sequences.fasta"
muscle_cline = MuscleCommandline(input=file, out=aligned_file)
muscle_cline()

# Read the aligned sequences
alignment = AlignIO.read(aligned_file, "fasta")

# Create a distance matrix from the alignment
calculator = DistanceCalculator('identity')
distance_matrix = calculator.get_distance(alignment)

# Construct the UPGMA tree
constructor = DistanceTreeConstructor()
upgma_tree = constructor.upgma(distance_matrix)

# Visualize the tree
Phylo.draw(upgma_tree)
```

muscle_cline() executes the Muscle command, aligning the sequences from the input FASTA file and saving the results in the output file.

calculator = DistanceCalculator('identity') creates a distance calculator using the identity (fraction of identical positions) to calculate the distance matrix from the alignment.

```python
from Bio import Phylo
from Bio.Align.Applications import MuscleCommandline
from Bio import AlignIO
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor, DistanceCalculator

file = "/home/charith/Desktop/multiple.fasta"

# Perform sequence alignment using MUSCLE
aligned_file = "/home/charith/Desktop/aligned_sequences.fasta"
muscle_cline = MuscleCommandline(input=file, out=aligned_file)
muscle_cline()

# Read the aligned sequences
alignment = AlignIO.read(aligned_file, "fasta")

# Create a distance matrix from the alignment
calculator = DistanceCalculator('identity')
distance_matrix = calculator.get_distance(alignment)

# Construct the NJ tree
constructor = DistanceTreeConstructor()
nj_tree = constructor.nj(distance_matrix)

# Visualize the tree
Phylo.draw(nj_tree)
```
